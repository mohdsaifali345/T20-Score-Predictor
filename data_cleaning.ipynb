```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames = []
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
```

```python
final_df = pd.DataFrame()
counter = 1
for file in tqdm(filenames):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        final_df = pd.concat([final_df, df], ignore_index=True)
        counter+=1
        
final_df
```

```python

matches = pickle.load(open('dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
import pandas as pd

delivery_data = []  # Initialize an empty list to store data

# Iterate through rows of the 'matches' DataFrame
for index, row in matches.iterrows():
    # Skip certain matches
    if count in [75, 108, 150, 180, 268, 360, 443, 458, 584, 748, 982, 1052, 1111, 1226, 1345]:
        count += 1
        continue
    
    # Extract data from the current match
    match_id = count
    teams = row['info.teams']
    batting_team = row['innings'][0]['1st innings']['team']
    city = row['info.city']
    venue = row['info.venue']
    
    # Extract delivery-level data
    for delivery in row['innings'][0]['1st innings']['deliveries']:
        ball = list(delivery.keys())[0]
        ball_data = delivery[ball]
        batsman = ball_data['batsman']
        bowler = ball_data['bowler']
        runs = ball_data['runs']['total']
        player_dismissed = ball_data.get('wicket', {}).get('player_out', '0')
        
        # Append data for each delivery to the list
        delivery_data.append({
            'match_id': match_id,
            'teams': teams,
            'batting_team': batting_team,
            'ball': ball,
            'batsman': batsman,
            'bowler': bowler,
            'runs': runs,
            'player_dismissed': player_dismissed,
            'city': city,
            'venue': venue
        })
    
    count += 1

# Create DataFrame from the collected delivery data
delivery_df = pd.DataFrame(delivery_data)
```

```python
delivery_df
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
